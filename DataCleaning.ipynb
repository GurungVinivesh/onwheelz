# **REPLACE 0 WITH NAN**

```python
#All_Columns
columns = list(data)
columns
```

```python
#Checking null values
data.isnull().sum()
```

```python
#calculating total zeroes in Glucose, BloodPressure, SkinThickness, Insulin and BMI
#Replace satement
(data[columns[1:6]] == 0).sum()
data[columns[1:6]] = data[columns[1:6]].replace(0,np.nan)
data
```

```python
#Again checkinig null values
data.isnull().sum()
```

# **DROPING THE EMPTY ROWS**

```python
#Before drop statement
data.shape
```

```python
#Drop Statement
data.dropna(inplace=True)
data
```
